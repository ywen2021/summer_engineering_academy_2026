## 1. What is a Large Language Model?

A **Large Language Model**, or **LLM**, is an AI model that learns patterns from a very large amount of text.

It can help with tasks such as:

- finishing a sentence
- answering questions
- summarizing text
- translating languages
- writing drafts
- explaining code
- brainstorming ideas

An LLM does **not** truly “understand” like a human. It predicts likely words based on patterns it learned during training.

## 3. Import the Tools

The main tool we will use is called `pipeline`.

A **pipeline** is a simple way to use an AI model without writing a lot of complicated code.

```python
from transformers import pipeline, set_seed
import textwrap

def print_wrapped(text, width=90):
    # Print long text in a more readable way.
    print(textwrap.fill(text, width=width))
```

## 4. First Example: Sentiment Analysis

**Sentiment analysis** means deciding whether a sentence sounds positive or negative.

For example:

- “I love this class!” → positive
- “This was confusing and frustrating.” → negative

Let’s try it.

```python
sentiment_model = pipeline("sentiment-analysis")

sentences = [
    "I am excited to learn about artificial intelligence!",
    "This homework is difficult and confusing.",
    "The movie was okay, but the ending was amazing."
]

results = sentiment_model(sentences)

for sentence, result in zip(sentences, results):
    print("Sentence:", sentence)
    print("Prediction:", result)
    print()
```

## 5. Text Generation with a Small LLM

Now we will use a small language model called **DistilGPT2**.

It is much smaller than tools like ChatGPT, so the output may be less impressive. But it is useful for learning because it can run in a notebook.

The model starts with a prompt and tries to continue it.

```python
generator = pipeline("text-generation", model="distilbert/distilgpt2")
set_seed(42)

prompt = "In the future, robots will help humans by"

output = generator(
    prompt,
    max_new_tokens=40,
    num_return_sequences=1,
    do_sample=True,
    temperature=0.8,
    pad_token_id=generator.tokenizer.eos_token_id
)

print_wrapped(output[0]["generated_text"])
```

### What Just Happened?

The model was given this beginning:

> In the future, robots will help humans by

Then it predicted words that could come next.

Important idea:

> The model is not copying one correct answer. It is generating a possible continuation.

## 7. What Are Tokens?

LLMs do not read text exactly the way humans do. They break text into smaller pieces called **tokens**.

A token can be:

- a word
- part of a word
- punctuation
- a space-like symbol

Let’s see how a tokenizer breaks a sentence into pieces.

```python
tokenizer = generator.tokenizer

sentence = "Artificial intelligence can help students learn in creative ways."

tokens = tokenizer.tokenize(sentence)
token_ids = tokenizer.convert_tokens_to_ids(tokens)

print("Original sentence:")
print(sentence)

print("\nTokens:")
print(tokens)

print("\nToken IDs:")
print(token_ids)

print("\nNumber of tokens:", len(tokens))
```

## 8. Mini Chatbot Function

The model we are using is not a real chat assistant, but we can still build a simple text-generation demo.

This function takes your message and asks the model to continue it.

```python
def simple_llm_response(user_message, max_new_tokens=60, temperature=0.8):
    prompt = "Student: " + user_message + "\nAI:"
    output = generator(
        prompt,
        max_new_tokens=max_new_tokens,
        num_return_sequences=1,
        do_sample=True,
        temperature=temperature,
        pad_token_id=generator.tokenizer.eos_token_id
    )
    response = output[0]["generated_text"]
    return response

message = "Can you give me one idea for using AI to help the environment?"
print_wrapped(simple_llm_response(message))
```
